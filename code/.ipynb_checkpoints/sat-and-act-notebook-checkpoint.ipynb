#### Fixing Incorrect Data Types
Defining a function to correct the incorrect 'Participation' datatypes and running it with the pandas .map() function.

```python
# To be used with the pandas .map function. Strips the '%' character from a passed string and returns the result.
def convert_participation(cell):
    return float(cell.replace('%', '')) / 100
```

```python
# Maps the 'convert_participation' function to the two 'Participation' columns in each dataframe.
sat_2017_df['Participation'] = sat_2017_df['Participation'].map(convert_participation)
act_2017_df['Participation'] = act_2017_df['Participation'].map(convert_participation)
```

```python
act_2017_df['Composite'] = act_2017_df['Composite'].map(lambda cell: float(cell.replace('x','')))
```

#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Custom function defined to manually calculate the standard deviation

```python
# Returns the standard deviation of a list 'x'
def standard_deviation(x):
    mean = sum(x) / len(x)
    variance = sum([(i - mean)**2 for i in x]) / len(x)
    return (variance ** 0.5)
```

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

```python
# Saves the calculated standard deviation to the 'sd' variable. Calls the 'to_dict' function on the final_df
# to translate the dataframe into a dictionary.
sd = {key: standard_deviation(value) for key, value in final_df.to_dict('list').items()}
sd
```

```python
# Comparing results of the manual standard deviation calculation to the numpy.std() function
{key: np.std(value) for key, value in final_df.to_dict('list').items()}
```

>The manually calculated standard deviations match up with the output from numpy's 'std' method, but are slightly off of the pandas 'describe' function. This can be explained by the rationale behind how 'describe' works - it assumes 1 degree of freedom whereas the manual and numpy methods default to 0.

#### Defining a custom function to subplot histograms

```python
# Function to create subplots of histograms based on a given list of columns, titles, and a figure size
def subplot_histograms(dataframe, list_of_columns, list_of_titles, figure_size):
    nrows = int(np.ceil(len(list_of_columns)/2)) # Makes sure there are enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols = 2, figsize = figure_size, 
                           gridspec_kw={'wspace': 0.15, 'hspace': 0.4})  # Gridspec_kw sets the whitespace
    ax = ax.ravel() # Ravel turns a matrix into a vector, which is easier to iterate
    
    for i, column in enumerate(list_of_columns): # Loops through all columns, plots a histogram and adds a title.
        ax[i].hist(dataframe[column])
        ax[i].set_title(f'{list_of_titles[i]}')
```

#### Plot and interpret histograms

- Participation rates for SAT & ACT
- Math scores for SAT & ACT
- Reading/verbal scores for SAT & ACT

```python
# Passes the dataframe, a list of columns, a list of titles, and a figure size to the subplotting function.
subplot_histograms(final_df, 
                   list_of_columns = ['2017_sat_participation', '2018_sat_participation',
                                      '2017_sat_reading',       '2018_sat_reading',
                                      '2017_sat_math',          '2018_sat_math',
                                      '2017_act_participation', '2018_act_participation',
                                      '2017_act_reading',       '2017_act_english',
                                      '2017_act_math'
                                     ], 
                   list_of_titles = ['Dist. of 2017 SAT Participation',  'Dist. of 2018 SAT Participation',
                                     'Dist. of 2017 SAT Reading Scores', 'Dist of 2018 SAT Reading Scores', 
                                     'Dist. of 2017 SAT Math Scores',    'Dist of 2018 SAT Math Scores',
                                     'Dist. of 2017 ACT Participation',  'Dist. of 2018 ACT Participation',
                                     'Dist. of 2017 ACT Reading Scores', 'Dist of 2017 ACT English Scores',
                                     'Dist. of 2017 ACT Math Scores'
                                    ],
                   figure_size = (12,18))
```

#### Defining a custom function to subplot scatter plots

```python
# Function to create subplots of histograms based on a given list of columns, titles, and a figure size
def subplot_scatterplots(dataframe, xcols, ycols, titles):
    nrows = int(np.ceil(len(xcols) / 2))
    fig, axes = plt.subplots(nrows = nrows, ncols = 2, figsize = (10,10),
                            gridspec_kw={'wspace': 0.25, 'hspace': 0.25})
    axes = axes.ravel()
    for i, column in enumerate(xcols):
        sns.regplot(x = dataframe[xcols[i]], 
                    y = dataframe[ycols[i]], 
                    data = dataframe,
                    fit_reg = False,
                    ax = axes[i]).set_title(titles[i])
```

```python
# Passes the dataframe, a list of columns, a list of titles, and a figure size to the subplotting function.
subplot_scatterplots(final_df, 
                     xcols = ['2017_act_math', 
                              '2017_act_reading',
                              '2017_act_total'
                             ], 
                     ycols = ['2017_sat_math', 
                              '2017_sat_reading',
                              '2017_sat_total'
                             ], 
                     titles = ['ACT vs SAT Math - 2017',
                               'ACT vs SAT Reading - 2017',
                               'ACT vs SAT Total - 2017'])
```
